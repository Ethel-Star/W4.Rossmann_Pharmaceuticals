```python
import numpy as np

# Checking for missing values in each dataset
missing_train = train_df.isnull().sum()
missing_test = test_df.isnull().sum()
missing_store = store_df.isnull().sum()

# Detecting outliers in numerical columns of the train and store datasets using Z-score
from scipy.stats import zscore

# Calculate Z-scores for numerical columns in train and store datasets
train_numeric_cols = train_df.select_dtypes(include=[np.number])
store_numeric_cols = store_df.select_dtypes(include=[np.number])

# Compute Z-scores and identify potential outliers (e.g., |Z| > 3)
train_outliers = (np.abs(zscore(train_numeric_cols)) > 3).sum()
store_outliers = (np.abs(zscore(store_numeric_cols)) > 3).sum()

(missing_train, missing_test, missing_store, train_outliers, store_outliers)
```

```python
# Handling missing values in the store dataset

# 1. Fill 'CompetitionDistance' missing values with the median, assuming competition is generally present
store_df['CompetitionDistance'].fillna(store_df['CompetitionDistance'].median(), inplace=True)

# 2. Fill 'CompetitionOpenSinceMonth' and 'CompetitionOpenSinceYear' missing values
# Assuming missing means no competition, fill with 0
store_df['CompetitionOpenSinceMonth'].fillna(0, inplace=True)
store_df['CompetitionOpenSinceYear'].fillna(0, inplace=True)

# 3. Fill 'Promo2SinceWeek', 'Promo2SinceYear', 'PromoInterval' missing values
# Assuming missing values mean no Promo2 participation
store_df['Promo2SinceWeek'].fillna(0, inplace=True)
store_df['Promo2SinceYear'].fillna(0, inplace=True)
store_df['PromoInterval'].fillna('None', inplace=True)

# Handling missing values in the test dataset
# Fill 'Open' with 1 (assuming the store was open if not specified)
test_df['Open'].fillna(1, inplace=True)

# Re-check missing values
missing_store_after = store_df.isnull().sum()
missing_test_after = test_df.isnull().sum()


missing_train = train_df.isnull().sum()
missing_test = test_df.isnull().sum()
missing_store = store_df.isnull().sum()
(missing_train, missing_test, missing_store, train_outliers, store_outliers)
```

```python
# Calculate the Pearson correlation between Sales and Customers
correlation = train_df[['Sales', 'Customers']].corr()

# Print the correlation
print("Correlation between Sales and Number of Customers:")
print(correlation)

# Plot the relationship between Sales and Customers
plt.figure(figsize=(10, 6))
sns.scatterplot(data=train_df, x='Customers', y='Sales')
plt.title('Scatter Plot of Sales vs. Number of Customers')
plt.xlabel('Number of Customers')
plt.ylabel('Sales')
plt.show()
```

```python
import pandas as pd
import statsmodels.api as sm

# Prepare data
# Ensure 'Sales' and 'Customers' columns are available and clean
X = train_df[['Customers']]  # Independent variable
y = train_df['Sales']        # Dependent variable

# Add a constant to the model (intercept)
X = sm.add_constant(X)

# Fit the OLS regression model
model = sm.OLS(y, X).fit()

# Print the model summary
print(model.summary())
```

Coefficients:
Intercept (const): 1077.7365. This represents the estimated sales when there are no customers. While this figure isn't practical, it helps anchor the regression line.

Customer Count: 7.4171. For each additional customer, sales are expected to increase by approximately 7.42 units, indicating a strong positive relationship between customer count and sales.

P-values:
Both the intercept and the customer coefficient have p-values of 0.000, indicating they are statistically significant. This means we can confidently assert that the number of customers is a valuable predictor of sales.

F-statistic (4.082e+06):
This high F-statistic suggests that at least one predictor variable (in this case, customers) is strongly associated with the dependent variable (sales).

Durbin-Watson (1.527):
This statistic tests for autocorrelation in the residuals. Values close to 2 indicate no autocorrelation; here, it suggests some positive autocorrelation, but not to a concerning degree.

Condition Number (1.33e+03):
A high condition number may indicate multicollinearity or numerical instability. However, this isn't a major concern here since the model includes only one predictor.

#### Promo Effectiveness on Existing vs New Customers

```python
# Creating categories for Customers
train_df['CustomerCategory'] = pd.qcut(train_df['Customers'], q=3, labels=['Small', 'Medium', 'Large'])

# Promo effectiveness on customer categories
promo_customer_sales = train_df.groupby(['Promo', 'CustomerCategory'], observed=False)['Sales'].mean().unstack()

# Plotting promo effectiveness on customer categories
promo_customer_sales.plot(kind='bar', stacked=True, figsize=(10, 6), color=['lightgreen', 'lightblue', 'salmon'])
plt.title('Promo Effectiveness on Sales (Small, Medium, Large Customers)')
plt.xlabel('Promo')
plt.ylabel('Average Sales')
plt.legend(title='Customer Category')
plt.show()
```
